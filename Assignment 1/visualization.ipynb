```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
import pandas as pd 
```

```python
df = pd.read_csv('A1_dataset_processed.csv')
stopwords = set(STOPWORDS)
```

```python
pos_text = " ".join(list(df[df.LABEL==1].NEW_TEXT.astype(str)))
neg_text = " ".join(list(df[df.LABEL==0].NEW_TEXT.astype(str)))
```

```python
wordcloud1=WordCloud(width = 800, height = 800,
                background_color ='white',
                # stopwords = stopwords,
                min_font_size = 10).generate(pos_text)
plt.figure(figsize=[10,8])
plt.axis("off")
plt.imshow(wordcloud1)
```

```python
wordcloud2=WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(neg_text)
plt.figure(figsize=[10,8])
plt.axis("off")
plt.imshow(wordcloud2)
```

## Observations

The occurence of positive sentiment words (such as love, thank, lol, haha, great, awesome etc.) is much higher in tweets with a positive sentiment score.

The occurence of negative sentiment words (such as miss, sad, bad, sick, bore, lose etc.) is much higher in tweets with a negative sentiment score.

However, there are exceptions. Eg- the word good appears in significantly high number in both negative and positive tweets as follows:
Positive sentence : 'What a good day'.<br>
Negative sentence : 'headache and not feeling good again'. <br>

Further both wordclouds have occurences of words that hold different sentimental meaning in different contexts. Eg- think appears in both positive and negative tweets as follows:<br>
Positive sentence: 'I think this voice thing is working out'.<br>
Negative sentnec : 'but they dont think of the consequences'.<br>

```python
import pandas as pd 
import numpy as np
import re
from IPython.display import display
```

```python
df = pd.read_csv('/content/A1_dataset.csv')
df.head()
```

```python
pos_df = df[df.LABEL==1]
neg_df = df[df.LABEL==0]
pos_txt = pos_df['TEXT']
neg_txt = neg_df['TEXT']
display('pos_df',pos_df.head(),'neg_df',neg_df.head())
```

URL Finder : This is to remove urls before word/sentence tokenization

```python
txt = pos_df['TEXT'].iloc[[27,54,71]]
pre_url = r'(?:https?://)|(?:www\.)'
url_set = r'[\w+=?/._\-:;&*^$#@~`!|]'
url_finder= re.compile(f'(?:{pre_url}){url_set}+|{url_set}+\.com{url_set}*')

# [(tok,i) for i,sen in enumerate(txt) for tok in url_finder.findall(sen)]
display([sen for sen in txt])
display([url_finder.sub('',sen) for sen in txt])
```

### PART A

Average Number of Sentences

```python
pos_sen=[tok for sen in pos_txt for tok in re.split(r'[!.?]| - ',url_finder.sub('url',sen)) if re.findall('\w',tok)]
neg_sen=[tok for sen in neg_txt for tok in re.split(r'[!.?]| - ',url_finder.sub('url',sen)) if re.findall('\w',tok)]
pos_tokens = [tok for sen in pos_sen for tok in re.split('\s+',sen) if re.search("\w",tok)]
neg_tokens = [tok for sen in neg_sen for tok in re.split('\s+',sen) if re.search("\w",tok)]

# pos_tokens = [tok for sen in pos_sen for tok in re.findall('\w+',sen)]
# neg_tokens = [tok for sen in neg_sen for tok in re.findall('\w+',sen)]


print("Average Number of Sentence in Positive Text:",len(pos_sen)/len(pos_txt))
print("Average Number of Sentence in Negative Text:",len(neg_sen)/len(neg_txt))
print()
print("Average Number of Tokens in Positive Text",len(pos_tokens)/len(pos_txt))
print("Average Number of Tokens in Negative Text",len(neg_tokens)/len(neg_txt))
```

Words starting with consonants and vowels

```python
consonant_set ='[bcdfghjklmnpqrstvwxyz]'
vowel_set = '[aeiou]'

pos_consonant = [tok for sen in pos_txt for tok in re.findall(r'\b'+f'{consonant_set}\w*',sen)]
neg_consonant = [tok for sen in neg_txt for tok in re.findall(r'\b'+f'{consonant_set}\w*',sen)]

pos_vowel = [tok for sen in pos_txt for tok in re.findall(r'\b'+f'{vowel_set}\w*',sen)]
neg_vowel = [tok for sen in neg_txt for tok in re.findall(r'\b'+f'{vowel_set}\w*',sen)]

print("Number of words starting with consonants in Postive Text: ",len(pos_consonant))
print("Number of words starting with consonants in Negative Text: ",len(neg_consonant))
print()
print("Number of words starting with vowels in Postive Text: ",len(pos_vowel))
print("Number of words starting with vowels in Negative Text: ",len(neg_vowel))
```

Lowercased tokens

```python
pos_lower_unique_tokens=set([tok.lower()  for tok in pos_tokens])
neg_lower_unique_tokens=set([tok.lower()  for tok in neg_tokens])
pos_unique_tokens=set(pos_tokens)
neg_unique_tokens=set(neg_tokens)

print("Number of unique tokens before lowercasing in positive text: ",len(pos_unique_tokens))
print("Number of unique tokens before lowercasing in negative text: ",len(neg_unique_tokens))
print()
print("Number of unique tokens after lowercasing in positive text: ",len(pos_lower_unique_tokens))
print("Number of unique tokens after lowercasing in negative text: ",len(neg_lower_unique_tokens))
```

```python
username_finder = re.compile('@\w+')

pos_usernames = [tok for sen in pos_txt for tok in username_finder.findall(sen)]
neg_usernames = [tok for sen in neg_txt for tok in username_finder.findall(sen)]

print("Number of usernames in positive text:", len(pos_usernames))
print("Number of usernames in negative text:", len(neg_usernames))

display('List of usernames in positive text',pos_usernames)
display('List of usernames in negative text',neg_usernames)
```

```python
pos_urls = [tok for sen in pos_txt for tok in url_finder.findall(sen)]
neg_urls = [tok for sen in neg_txt for tok in url_finder.findall(sen)]

print("Number of urls in positive text:", len(pos_urls))
print("Number of urls in negative text:", len(neg_urls))

display('List of urls in positive text',pos_urls)
display('List of urls in negative text',neg_urls)
```

Tweets each day of the week

```python
copy_df=df.copy(deep=True)
copy_df['Day'] = df['DATE_TIME'].str.split(' ').str[0]
copy_df.head()
```

```python
# def count_tweets(row,day,count=0):
#   if row==day:
#     count+=1
#   return count
```

```python
#find unique days in the dataset
days=copy_df['Day'].unique()

#function to detect if tweet exist on particular day
def count_tweets(row,day,count=0):
  match = re.search(r'\b'+f'{day}'+r'\b',row)
  if match:
    count+=1
  return count
  
####
print('Number of tweets on each day of the week')
for day in days:
  count=len(copy_df[copy_df.DATE_TIME.apply(lambda row:count_tweets(row,day))==1].index)
  
  print(f'{day} : {count}')
```

### PART B

```python
word, label = input("Enter word and label: ").split(' ')
label=int(label)
sentiment = 'positive' if label==1 else 'negative'
```

Total Number of occurences of word and sentences containing that word

```python
if label==1:
  word_count=[tok for tok in pos_tokens if re.search(r'\b'+ word + r'\b', tok, re.IGNORECASE)]
  sent_count= [sen for sen in pos_sen if re.search(r'\b'+ word + r'\b', sen, re.IGNORECASE)]

else:
  word_count=[tok for tok in neg_tokens if re.search(r'\b'+ word + r'\b', tok, re.IGNORECASE)]
  sent_count= [sen for sen in neg_sen if re.search(r'\b'+ word + r'\b', sen, re.IGNORECASE)]

print(f"Total Number of occurences of the word '{word}' in {sentiment} text: {len(word_count)}")
print(f"Total Number of sentences containing the word '{word}' in {sentiment} text: {len(sent_count)}")
```

Number of Sentences starting with a word

```python
if label==1:
    start_sent = [sen for sen in pos_sen if re.search('^\s*'+ word + r'\b', sen, re.IGNORECASE)]
else:
    start_sent = [sen for sen in neg_sen if re.search('^\s*'+ word + r'\b', sen, re.IGNORECASE)]

print(f"Number of sentences starting with the word {word} in {sentiment} text : {len(start_sent)}")
```

Number of Sentences ending with a word

```python
if label==1:
  end_sent = [sen for sen in pos_sen if re.search(r'\b'+ word + '\s*$', sen, re.IGNORECASE)]
else:
  end_sent = [sen for sen in neg_sen if re.search(r'\b'+ word + '\s*$', sen, re.IGNORECASE)]

print(f"Number of sentences ending with the word {word} in {sentiment} text : {len(end_sent)}")
```
